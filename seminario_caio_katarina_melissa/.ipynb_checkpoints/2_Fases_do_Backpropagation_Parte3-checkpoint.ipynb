# Fases do Backpropagation - Parte 3

## Introdução

Neste notebook, exploraremos a última fase do algoritmo de Backpropagation: a Atualização dos Pesos. Esta fase é onde o aprendizado efetivamente ocorre, pois os pesos da rede neural são ajustados para minimizar o erro.

```python
# Importação das bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib.colors as mcolors
from matplotlib import cm
import math
```

## 4. Atualização dos Pesos

Após calcular os gradientes do erro em relação a cada peso durante a fase de Backpropagation, a fase final consiste em atualizar os pesos da rede para reduzir o erro. Esta atualização é geralmente realizada usando o algoritmo do gradiente descendente ou suas variantes.

### Algoritmo do Gradiente Descendente

O gradiente descendente é um algoritmo de otimização que busca encontrar o mínimo de uma função ajustando iterativamente seus parâmetros na direção oposta ao gradiente da função.

No contexto do treinamento de redes neurais, a função a ser minimizada é a função de erro, e os parâmetros são os pesos e vieses da rede. A regra de atualização básica do gradiente descendente é:

$w_{ji}^l = w_{ji}^l - \eta \frac{\partial E}{\partial w_{ji}^l}$

Onde:
- $w_{ji}^l$ é o peso da conexão entre o neurônio $i$ na camada $l-1$ e o neurônio $j$ na camada $l$
- $\eta$ (eta) é a taxa de aprendizado, um hiperparâmetro que controla o tamanho do passo na direção do gradiente
- $\frac{\partial E}{\partial w_{ji}^l}$ é o gradiente do erro em relação ao peso $w_{ji}^l$

De forma similar, os vieses são atualizados usando:

$b_j^l = b_j^l - \eta \frac{\partial E}{\partial b_j^l}$

### Taxa de Aprendizado

A taxa de aprendizado $\eta$ é um hiperparâmetro crucial que controla o quão rápido os pesos são ajustados. Ela tem um impacto significativo no processo de treinamento:

- **Taxa de aprendizado muito alta**: Pode causar oscilações ou divergência, impedindo que o algoritmo encontre o mínimo da função de erro.
- **Taxa de aprendizado muito baixa**: Pode resultar em um treinamento muito lento, exigindo muitas épocas para convergir.

Encontrar a taxa de aprendizado ideal geralmente requer experimentação, embora existam técnicas adaptativas que ajustam automaticamente a taxa de aprendizado durante o treinamento.

```python
# Implementação da atualização de pesos com gradiente descendente
def update_weights(weights, biases, nabla_w, nabla_b, learning_rate):
    """Atualiza os pesos e vieses usando o gradiente descendente
    
    Args:
        weights: Lista de matrizes de pesos para cada camada
        biases: Lista de vetores de viés para cada camada
        nabla_w: Gradientes para os pesos
        nabla_b: Gradientes para os vieses
        learning_rate: Taxa de aprendizado
        
    Returns:
        weights: Pesos atualizados
        biases: Vieses atualizados
    """
    # Atualizar cada peso e viés usando a regra do gradiente descendente
    weights = [w - learning_rate * nw for w, nw in zip(weights, nabla_w)]
    biases = [b - learning_rate * nb for b, nb in zip(biases, nabla_b)]
    
    return weights, biases

# Exemplo de uso
# Definir uma rede neural simples com 2 entradas, 3 neurônios na camada oculta e 1 saída
weights = [np.random.randn(3, 2), np.random.randn(1, 3)]
biases = [np.random.randn(3, 1), np.random.randn(1, 1)]

# Dados de exemplo
x = np.array([[0.5], [0.3]])  # Entrada
y = np.array([[0.7]])  # Saída desejada

# Definir taxa de aprendizado
learning_rate = 0.1

# Calcular os gradientes usando backpropagation
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))

def backpropagation(x, y, weights, biases):
    # Inicializar gradientes
    nabla_w = [np.zeros(w.shape) for w in weights]
    nabla_b = [np.zeros(b.shape) for b in biases]
    
    # Feedforward
    activation = x
    activations = [x]
    zs = []
    
    for w, b in zip(weights, biases):
        z = np.dot(w, activation) + b
        zs.append(z)
        activation = sigmoid(z)
        activations.append(activation)
    
    # Backpropagation
    delta = (activations[-1] - y) * sigmoid_derivative(zs[-1])
    nabla_b[-1] = delta
    nabla_w[-1] = np.dot(delta, activations[-2].transpose())
    
    for l in range(2, len(weights) + 1):
        delta = np.dot(weights[-l+1].transpose(), delta) * sigmoid_derivative(zs[-l])
        nabla_b[-l] = delta
        nabla_w[-l] = np.dot(delta, activations[-l-1].transpose())
    
    return nabla_w, nabla_b, activations[-1]

# Treinar a rede por algumas épocas
num_epochs = 10
errors = []

print("Treinamento da rede neural por", num_epochs, "épocas:")
for epoch in range(num_epochs):
    # Calcular gradientes
    nabla_w, nabla_b, output = backpropagation(x, y, weights, biases)
    
    # Calcular erro
    error = np.mean(np.square(y - output))
    errors.append(error)
    
    # Atualizar pesos
    weights, biases = update_weights(weights, biases, nabla_w, nabla_b, learning_rate)
    
    print(f"Época {epoch+1}: Erro = {error:.6f}, Saída = {output[0][0]:.6f}")

# Plotar a evolução do erro
plt.figure(figsize=(10, 6))
plt.plot(range(1, num_epochs+1), errors, 'o-', linewidth=2)
plt.title('Evolução do Erro Durante o Treinamento')
plt.xlabel('Época')
plt.ylabel('Erro Quadrático Médio')
plt.grid(True)
plt.show()
```

### Variantes do Gradiente Descendente

Existem várias variantes do algoritmo do gradiente descendente que podem melhorar a convergência e o desempenho:

1. **Gradiente Descendente Estocástico (SGD)**: Atualiza os pesos usando um único exemplo de treinamento por vez, em vez de todo o conjunto de dados.

2. **Gradiente Descendente com Momentum**: Adiciona um termo de momentum que ajuda a acelerar a convergência e evitar mínimos locais.
   
   $v = \gamma v - \eta \nabla E$
   
   $w = w + v$
   
   Onde $v$ é o vetor de velocidade e $\gamma$ é o coeficiente de momentum.

3. **RMSProp**: Adapta a taxa de aprendizado para cada parâmetro com base na magnitude dos gradientes recentes.

4. **Adam**: Combina as ideias do momentum e do RMSProp para uma adaptação mais eficiente da taxa de aprendizado.

### Visualização da Atualização dos Pesos

```python
def create_weight_update_visualization():
    """Cria uma visualização para o processo de atualização de pesos"""
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 6)
    ax.axis('off')
    
    # Posições dos neurônios
    input_pos = [(2, i) for i in range(1, 5)]
    hidden_pos = [(5, i) for i in range(1, 6)]
    output_pos = [(8, i) for i in range(2, 5)]
    
    # Desenhar neurônios
    input_neurons = [plt.Circle(pos, 0.3, fc='lightblue', ec='blue') for pos in input_pos]
    hidden_neurons = [plt.Circle(pos, 0.3, fc='lightgreen', ec='green') for pos in hidden_pos]
    output_neurons = [plt.Circle(pos, 0.3, fc='salmon', ec='red') for pos in output_pos]
    
    # Adicionar neurônios ao gráfico
    for neuron in input_neurons + hidden_neurons + output_neurons:
        ax.add_patch(neuron)
    
    # Adicionar texto para identificar as camadas
    ax.text(2, 0.3, 'Camada de Entrada', ha='center', fontsize=12)
    ax.text(5, 0.3, 'Camada Oculta', ha='center', fontsize=12)
    ax.text(8, 0.3, 'Camada de Saída', ha='center', fontsize=12)
    
    # Desenhar conexões entre neurônios
    connections = []
    
    # Conexões da camada de entrada para a camada oculta
    for i, pos_in in enumerate(input_pos):
        for j, pos_hid in enumerate(hidden_pos):
            line, = ax.plot([pos_in[0]+0.3, pos_hid[0]-0.3], [pos_in[1], pos_hid[1]], 'k-', alpha=0.2, lw=1)
            connections.append((line, 'input-hidden', i, j))
    
    # Conexões da camada oculta para a camada de saída
    for i, pos_hid in enumerate(hidden_pos):
        for j, pos_out in enumerate(output_pos):
            line, = ax.plot([pos_hid[0]+0.3, pos_out[0]-0.3], [pos_hid[1], pos_out[1]], 'k-', alpha=0.2, lw=1)
            connections.append((line, 'hidden-output', i, j))
    
    # Adicionar símbolos de peso nas conexões
    weight_symbols = []
    for i, (line, layer, src, dst) in enumerate(connections):
        # Calcular posição do símbolo (no meio da conexão)
        x1, y1 = line.get_xdata()
        x_mid = (x1[0] + x1[1]) / 2
        y_mid = (y1[0] + y1[1]) / 2
        
        # Adicionar símbolo de peso
        weight = ax.text(x_mid, y_mid, "w", color='black', fontsize=10, ha='center', va='center', alpha=0)
        weight_symbols.append((weight, layer, src, dst))
    
    # Adicionar equação de atualização de peso
    equation = ax.text(5, 5.5, "", fontsize=14, ha='center', va='center', bbox=dict(facecolor='white', alpha=0.7))
    
    # Adicionar título
    title = ax.set_title('Fase de Atualização dos Pesos', fontsize=16)
    
    return fig, ax, connections, input_neurons, hidden_neurons, output_neurons, weight_symbols, equation, title

# Criar a visualização inicial
fig, ax, connections, input_neurons, hidden_neurons, output_neurons, weight_symbols, equation, title = create_weight_update_visualization()

# Função para animar o processo de atualização de pesos
def animate_weight_update(i):
    # Resetar cores e opacidade
    for neuron in input_neurons + hidden_neurons + output_neurons:
        neuron.set_alpha(0.7)
    
    for conn, _, _, _ in connections:
        conn.set_alpha(0.2)
        conn.set_color('k')
        conn.set_linewidth(1)
    
    for weight, _, _, _ in weight_symbols:
        weight.set_alpha(0)
    
    equation.set_alpha(0)
    
    # Fase 1: Introdução à atualização de pesos
    if i < 10:
        title.set_text('Fase de Atualização dos Pesos')
        equation.set_text("w = w - η ∇E")
        equation.set_alpha(1.0)
    
    # Fase 2: Atualização dos pesos entre a camada oculta e a camada de saída
    elif i < 25:
        title.set_text('Atualização dos Pesos: Camada Oculta → Camada de Saída')
        progress = (i - 10) / 14
        
        # Destacar neurônios relevantes
        for neuron in hidden_neurons:
            neuron.set_facecolor('green')
            neuron.set_alpha(1.0)
        for neuron in output_neurons:
            neuron.set_facecolor('red')
            neuron.set_alpha(1.0)
        
        # Destacar conexões e mostrar símbolos de peso
        for j, (conn, layer, src, dst) in enumerate(connections):
            if layer == 'hidden-output' and j % 3 <= (i - 10) / 5:
                conn.set_alpha(0.9)
                conn.set_color('gold')
                conn.set_linewidth(3)
        
        for j, (weight, layer, src, dst) in enumerate(weight_symbols):
            if layer == 'hidden-output' and j % 3 <= (i - 10) / 5:
                weight.set_alpha(1.0)
                weight.set_text("w - η∇E")
                weight.set_color('darkred')
        
        # Mostrar equação de atualização
        equation.set_text("wⱼᵏ = wⱼᵏ - η · δᵏ · aⱼ")
        equation.set_alpha(1.0)
    
    # Fase 3: Atualização dos pesos entre a camada de entrada e a camada oculta
    else:
        title.set_text('Atualização dos Pesos: Camada de Entrada → Camada Oculta')
        progress = (i - 25) / 14
        
        # Destacar neurônios relevantes
        for neuron in input_neurons:
            neuron.set_facecolor('blue')
            neuron.set_alpha(1.0)
        for neuron in hidden_neurons:
            neuron.set_facecolor('green')
            neuron.set_alpha(1.0)
        
        # Destacar conexões e mostrar símbolos de peso
        for j, (conn, layer, src, dst) in enumerate(connections):
            if layer == 'hidden-output':
                conn.set_alpha(0.4)
                conn.set_color('gold')
                conn.set_linewidth(1)
            elif layer == 'input-hidden' and j % 5 <= (i - 25) / 5:
                conn.set_alpha(0.9)
                conn.set_color('gold')
                conn.set_linewidth(3)
        
        for j, (weight, layer, src, dst) in enumerate(weight_symbols):
            if layer == 'hidden-output':
                weight.set_alpha(0.5)
                weight.set_text("w'")
                weight.set_color('darkred')
            elif layer == 'input-hidden' and j % 5 <= (i - 25) / 5:
                weight.set_alpha(1.0)
                weight.set_text("w - η∇E")
                weight.set_color('darkblue')
        
        # Mostrar equação de atualização
        equation.set_text("wᵢⱼ = wᵢⱼ - η · δⱼ · aᵢ")
        equation.set_alpha(1.0)
    
    return input_neurons + hidden_neurons + output_neurons + [conn[0] for conn in connections] + [weight for weight, _, _, _ in weight_symbols] + [equation, title]

# Criar a animação
ani = FuncAnimation(fig, animate_weight_update, frames=40, interval=150, blit=True)

# Exibir a animação
HTML(ani.to_jshtml())
```

### Impacto da Taxa de Aprendizado

```python
# Visualização do impacto da taxa de aprendizado
def plot_learning_rate_impact():
    """Plota o impacto de diferentes taxas de aprendizado na convergência"""
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Função de erro (parábola simples para ilustração)
    x = np.linspace(-2, 2, 1000)
    y = x**2
    
    # Plotar a função de erro
    ax.plot(x, y, 'k-', linewidth=2, label='Função de Erro')
    
    # Ponto inicial
    start_x = 1.5
    start_y = start_x**2
    ax.scatter([start_x], [start_y], color='black', s=100, label='Ponto Inicial')
    
    # Ponto mínimo
    min_x = 0
    min_y = min_x**2
    ax.scatter([min_x], [min_y], color='red', s=100, label='Mínimo Global')
    
    # Simular diferentes taxas de aprendizado
    learning_rates = [0.1, 0.5, 1.1]  # Baixa, média, alta
    colors = ['blue', 'green', 'orange']
    labels = ['Taxa Baixa (η=0.1)', 'Taxa Média (η=0.5)', 'Taxa Alta (η=1.1)']
    
    for lr, color, label in zip(learning_rates, colors, labels):
        # Simular o gradiente descendente
        x_path = [start_x]
        y_path = [start_y]
        
        current_x = start_x
        for _ in range(10):  # 10 iterações
            # Gradiente da função x^2 é 2x
            gradient = 2 * current_x
            # Atualização do gradiente descendente
            current_x = current_x - lr * gradient
            current_y = current_x**2
            
            x_path.append(current_x)
            y_path.append(current_y)
        
        # Plotar o caminho
        ax.plot(x_path, y_path, 'o-', color=color, linewidth=2, markersize=8, label=label)
    
    # Configurar o gráfico
    ax.set_xlim(-2, 2)
    ax.set_ylim(-0.5, 3)
    ax.set_xlabel('Peso', fontsize=12)
    ax.set_ylabel('Erro', fontsize=12)
    ax.set_title('Impacto da Taxa de Aprendizado na Convergência', fontsize=16)
    ax.legend(fontsize=10)
    ax.grid(True)
    
    plt.tight_layout()
    plt.show()

plot_learning_rate_impact()
```

## Conclusão sobre as Fases do Backpropagation

Nesta série de notebooks, exploramos as quatro fases principais do algoritmo de Backpropagation:

1. **Feedforward (Propagação Direta)**: Os dados de entrada são propagados através da rede para gerar uma saída.
2. **Cálculo do Erro**: A saída da rede é comparada com o valor desejado para determinar o erro.
3. **Backpropagation (Propagação Reversa)**: O erro é propagado de volta através da rede para calcular os gradientes.
4. **Atualização dos Pesos**: Os pesos da rede são ajustados na direção oposta ao gradiente para minimizar o erro.

Estas quatro fases trabalham em conjunto para permitir que a rede neural aprenda a partir dos dados de treinamento. O processo é repetido para múltiplos exemplos de treinamento e por várias épocas até que a rede atinja um desempenho satisfatório.

O Backpropagation é a base do treinamento de redes neurais modernas e, apesar de sua simplicidade conceitual, é incrivelmente poderoso quando combinado com arquiteturas de rede adequadas e técnicas de otimização avançadas.
